```python
import pandas as pd
```

```python
#1.Comparing elements of two pandas series
s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two series
comparison = s1 == s2

# Printing the comparison result
print(comparison)

```

```python
#2.Pandas program that performs addition,subtraction,multiplication,and division on two pandas series
a = pd.Series([2, 4, 6, 8, 14])
b = pd.Series([1, 3, 5, 7, 9])

# Perform addition
add= a + b

# Perform subtraction
sub = a - b

# Perform multiplication
mult =a * b

# Perform division
div = a / b

# Print the results
print("Addition:")
print(add)

print("\nSubtraction:")
print(sub)

print("\nMultiplication:")
print(mult)

print("\nDivision:")
print(div)
```

```python
#3.3. Pandas program to convert a dictionary to a Pandas series.
#  dictionary
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}

# Convert the dictionary to a Pandas series
series = pd.Series(dictionary1)

# Print the series
print(series)
```

```python
#4.converting series to an array
import numpy as np

#  the sample series
ser= pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

# Convert the series to an array
array = np.array(ser)

# Print the array
print(array)

```

```python
#5.Aprogram to display the most frequent value in the given series and replacing everything else as ‘Other’ in the series
#loading the dataset
df= pd.read_csv("AfricaCupofNationsMatches.csv")
Homes=df['HomeTeamGoals']
Home=pd.Series(Homes)
most_frequent=Home.mode()[0]
print("The most frequent value is:",most_frequent)
#Replacing the most frequent value with 'Other'
Home[Home != most_frequent] = 'Other'
Home
```

```python
#9.A Pandas program to sort the DataFrame first by ‘HomeTeam’ in ascending order, then by ‘HomeTeamScores’ in descending order.
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])

# Printing the sorted DataFrame
sorted_df
```

```python
#10. A Pandas program to get list from DataFrame column headers.
column_headers = df.columns.tolist()

# Print the list of column headers
column_headers
```

```python
#11.A Pandas program to append a column of your choice to your DataFrame.
Team_points=[]
for i in range(0,df.shape[0]):
    Team_points.append(i)
df['Points']=Team_points
df
```

```python
#12.A Pandas program to add 2 rows to the DataFrame.
new_rows = {
    'Year': [2023, 2023],
    'Date': ['10-Oct-23', '15-Oct-23'],
    'Time': [None, None],
    'HomeTeam': ['Germany', 'France'],
    'AwayTeam': ['Spain', 'Italy'],
    'HomeTeamGoals': [2, 3],
    'AwayTeamGoals': [1, 2],
    'Stage': ['Friendly', 'Friendly'],
    'SpecialWinConditions': [None, None],
    'Stadium': ['Stadium X', 'Stadium Y'],
    'City': ['City X', 'City Y'],
    'Attendance': [25000, 30000]
}

# Create a new DataFrame with the new rows
new_rows_df = pd.DataFrame(new_rows)

# Append the new DataFrame to the original DataFrame
df = df.append(new_rows_df, ignore_index=True)

# Print the updated DataFrame
df
```

```python
#13.changing the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')

# Print the updated DataFrame
print(df)
```

```python
#14.Reseting the index
df = df.reset_index(drop=True)

# Printing the updated DataFrame
df
```

```python
#15.checking whether the ‘Stadium’ column is present in the DataFrame or not.
if 'Stadium' in df.columns:
    print("The 'Stadium' column is present in the DataFrame.")
else:
    print("The 'Stadium' column is not present in the DataFrame.")
```

```python
#16.converting a datatype of 'AwayTeamGoals' into float
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)

# Print the updated DataFrame
print(df)
```

```python
#17. Remove the last 10 rows from the DataFrame
df = df.iloc[:-10]

# Printing the updated DataFrame
print(df)
```

```python
#18.A Pandas program to iterate over rows in the DataFrame
for index, row in df.iterrows():
    # Access row values using row
    print(f"Row {index}:")
    print(f"HomeTeam: {row['HomeTeam']}")
    print(f"AwayTeam: {row['AwayTeam']}")
    print(f"HomeTeamGoals: {row['HomeTeamGoals']}")
    print(f"AwayTeamGoals: {row['AwayTeamGoals']}")
    print(f"Stage: {row['Stage']}")
    print(f"Stadium: {row['Stadium']}")
    print(f"City: {row['City']}")
    print(f"Attendance: {row['Attendance']}")
    print()  
```

```python
#19.changing the order of dataframe columns
new_order = ['AwayTeam', 'HomeTeam', 'AwayTeamGoals', 'HomeTeamGoals', 'Stage', 'Stadium', 'City', 'Attendance']

# Reorder the DataFrame columns
df = df[new_order]

# Print the updated DataFrame
df
```

```python
#20.A Pandas program to delete DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.
df = df[df['HomeTeamGoals'] != 0]

# Print the new DataFrame
print(df)
```
